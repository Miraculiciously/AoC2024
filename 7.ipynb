```python
with open('7.txt', 'r') as file:
    lines = [line.strip() for line in file.readlines()]

lines
```

```python
puzzles = [ [int(line.split(":")[0]), [int(x) for x in line.split(": ")[1].split()]]  for line in lines ]
puzzles
```

```python
from itertools import product

def operator_combinations(length):
    return [list(comb) for comb in product(["+", "*"], repeat=length)]

operator_combinations(2)
```

```python
def eval_left_to_right(digits, operators):
    res = digits[0]

    assert(len(digits) -  1 == len(operators))
    
    for i, op in enumerate(operators):
        if op == "*":
            res *= digits[i + 1]  # Multiply by the next digit
        else: # op == "+"
            res += digits[i + 1]  # Add the next digit
    return res

test_digits = [2, 3, 4]
test_operators = ["+", "*"]

eval_left_to_right(test_digits, test_operators) # returns 20 although 2+3*4 should be 14, so does what it should
```

```python
# single instance evaluation through list comprehension over operators
[ eval_left_to_right(test_digits, test_operator) for test_operator in operator_combinations(len(test_digits) - 1) ]
```

```python
total = 0
for puzzle in puzzles:
    solution = puzzle[0]
    digit_list = puzzle[1]
    operators = operator_combinations(len(digit_list) - 1)
    if any(eval_left_to_right(digit_list, ops) == solution for ops in operators):
        total += solution
total
```

# 2

```python
def operator_combinations(length):
    return [list(comb) for comb in product(["+", "*", '||'], repeat=length)]

operator_combinations(2)
```

```python
def eval_left_to_right(digits, operators):
    res = digits[0]

    assert(len(digits) -  1 == len(operators))
    
    for i, op in enumerate(operators):
        if op == "*":
            res *= digits[i + 1]  # Multiply by the next digit
        elif op == "+":
            res += digits[i + 1]  # Add the next digit
        elif op == "||":
            res = int(f"{res}{digits[i + 1]}")
        else:
            raise ValueError("Oh no")
    return res

test_digits = [2, 3, 4, 5]
test_operators = ["+", "*", "||"]

eval_left_to_right(test_digits, test_operators) # returns 205 although 2+3*4 should be 14 so that would be 145
```

```python
from tqdm import tqdm

total = 0
for puzzle in tqdm(puzzles):
    solution = puzzle[0]
    digit_list = puzzle[1]
    operators = operator_combinations(len(digit_list) - 1)
    if any(eval_left_to_right(digit_list, ops) == solution for ops in operators):
        total += solution
total
```
